Can an algorithmic decision-making process ever be completely fair, even when it is designed to be? Black-box systems rule the world. People provide data, the system produces an output, and many pay little to no attention to how the system arrived at the answer. For most systems, the outputs fall somewhere on the spectrum of "that's weird" to "how did it know that, that's creepy". But in the case of immigration, the results can be dream-shattering or life-threatening.

Since March 2005, the US Department of Labor has used an electronic labor certification system known as PERM to reduce labor certification times (https://www.dol.gov/agencies/eta/foreign-labor/news/archive). With any electronic approval system, there would be concerns of potential bias. These concerns are only heightened when dealing with a controversial policy topic such as immigration.

Our team was asked to conduct a preliminary analysis of data from the PERM system from the Office of Foreign Labor Certification (https://www.dol.gov/agencies/eta/foreign-labor/performance#dis). Based on the 2020 disclosure file, we reviewed the feature set and performed exploratory analysis on the data to draw insights into the program and system. This system, which is designed to conduct a screening to determine whether someone is certified for the visa they applied for, is not the final determining factor, but instead an input used to further filter out ineligible applicants. 

### Import Packages, Read and Prep Data

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

sns.set_style('white')
#plt.xkcd(); # Yes, you can set the theme to xkcd cartoon style
```

```python
df = pd.read_excel(r"analysis and data/PERM_FY2020_Q3.xlsx")

# filter case statuses that are not withdrawn as including them could lead to misleading insights
df = df[df.CASE_STATUS != 'Withdrawn']

# create binary variables between cases that are denied vs certified
# denied cases will have a binary variable of 1 while certified cases will have a binary variable of 0
case_status = []
case_status_binary = []
for i in df.CASE_STATUS:
    if i == 'Denied':
        case_status.append('Denied')
        case_status_binary.append(1)
    else:
        case_status.append('Certified')
        case_status_binary.append(0)
        
df['CASE_STATUS'] = case_status
df['CASE_STATUS_BINARY'] = case_status_binary
```

### Looking at the pay distribution

```python
#Drop Null values prior to annualizing wages
df_temp = df[df['PW_UNIT_OF_PAY'].notna()]

#Annualize wage
counter = 0
annualized_wage = list()

for i in df_temp['PW_UNIT_OF_PAY']:
    if i == 'Year':
        annualized_wage.append(df_temp['PW_WAGE'].iloc[counter,])
    elif i == 'Hour':
        annualized_wage.append(df_temp['PW_WAGE'].iloc[counter,] * 2080)
    elif i == 'Week':
        annualized_wage.append(df_temp['PW_WAGE'].iloc[counter,] * 52)
    elif i == 'Month':
        annualized_wage.append(df_temp['PW_WAGE'].iloc[counter,] * 12)
    elif i == 'Bi-Weekly':
        annualized_wage.append(df_temp['PW_WAGE'].iloc[counter,] * 26)
#    else:
#       annualized_wage.append('0')
    
    counter = counter + 1

#print(annualized_wage)
df_temp['Annualized_Wage'] =  annualized_wage


#Create dataframe stripping out Null values in Education column
df1 = df_temp[df_temp['FOREIGN_WORKER_EDUCATION'].notna()]

#Analyze quick metrics on column
pd.DataFrame(df1['Annualized_Wage'].describe())
```

```python
#Analyze histogram of annualized wages for certified case outcomes
plt.figure(figsize = (15,5))
ax = df_temp.loc[df_temp['CASE_STATUS'] == 'Certified']['Annualized_Wage'].hist(bins = 25, range = [0,300000])
ax.xaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('${x:,.0f}'))

min_ylim, max_ylim = plt.ylim()
ax.axvline(df_temp.loc[df_temp['CASE_STATUS'] == 'Certified']['Annualized_Wage'].mean(), 
           color='k', linestyle='solid', linewidth=2)
plt.text(df_temp.loc[df_temp['CASE_STATUS'] == 'Certified']['Annualized_Wage'].mean()*.6, 
         max_ylim*0.975, 'Mean: ${:,.0f}'.format(df_temp.loc[df_temp['CASE_STATUS'] == 'Certified']['Annualized_Wage'].mean()))

plt.axvline(df_temp.loc[df_temp['CASE_STATUS'] == 'Certified']['Annualized_Wage'].median(), 
            color='k', linestyle='dashed', linewidth=2)
plt.text(df_temp.loc[df_temp['CASE_STATUS'] == 'Certified']['Annualized_Wage'].median()*1.1, 
         max_ylim*0.975, 'Median: ${:,.0f}'.format(df_temp.loc[df['CASE_STATUS'] == 'Certified']['Annualized_Wage'].median()))
```

### Analysis

To further bolster our case, while both the denied and certified wordclouds are dominated by technical and management positions, the denied cases have a noticably larger concentration of non-technical jobs, such as teachers, truck drivers and janitors.

Based on these observations, it is clear that the PERM system shows favor towards high skill and technical jobs, where there would be an expectation of high demand and low supply. That is in line with the current immigration policies, so there is definitely evidence of "by-design" bias. But by-design bias is not as much of a concern unintended bias.

## Unintended Bias

Unintended bias is where the design of the system or process results in discrimination or preference towards a particular group which wasn't an intention or objective of the design. In other words, it is an unwanted side effect. Some of the observations in the data set point towards inherent biases. For instance, when comparing to certified cases, denied cases had 10 times the rate of occurence of not having a Sunday newspaper in the area. This suggests skew against more rural areas.

```python
pd.crosstab(df["CASE_STATUS_BINARY"],
            df["SUNDAY_EDITION_NEWSPAPER"], 
            normalize='index')
```

Also of interest, denied cases were 11% less likely to have the bargaining rep notified than accepted cases, and had more than 25 times the rate of occurence of notices not being posted at the worksite, suggesting that workers who depend upon receiving information through these channels are less likely to have their application approved. This seems to align to the approval concentration in high skill jobs, but presents concerns that these workers who are otherwise eligible may not be getting all the information they need to navigate the process.

```python
pd.crosstab(df["CASE_STATUS_BINARY"],
            df["BARGAINING_REP_NOTIFIED"], 
            normalize='index')
```

```python
pd.crosstab(df["CASE_STATUS_BINARY"],
            df["POSTED_NOTICE_AT_WORKSITE"],
            normalize='index')
```

Denial rates also appear to be much higher amongst contries from Africa and South/Central/Caribbean America, suggesting potential for unitended bias. Conversely, India and China have extremely high approval numbers, and those numbers do not appear to be entirely explainable by their large populations, especially considering Inda has 4 times as many approved applications as China. These numbers appear largely consistent between the worker's birth country and country of citizenship.

```python
plt.figure(figsize=(15,12))
pd.crosstab(df["COUNTRY_OF_CITIZENSHIP"],df["CASE_STATUS_BINARY"],
            normalize='index')[1].sort_values().tail(15).plot.barh()
```

EWI stands for Entry Without Inspection, and means the applicant entered the country without inspection by an immgration or border patrol officer. 

## Conclusion

Based on the above data, our group has concerns that the existing process has unintended biases that should be further explored through feature engineering and data mining. There is data to suggest that the decision-making of the program is biased against emerging nations in ways that were unintended. Specifically, we would look for further studies to find areas for policy, system and process enhancement that could maintain the program's objectives while increasing the approval rates for emerging nations in Africa and South/Central/Caribbean America.
